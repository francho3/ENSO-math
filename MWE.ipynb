```python
import xarray as xr

from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'retina'
%matplotlib inline

n34_url = 'https://iridl.ldeo.columbia.edu/SOURCES/.NOAA/.NCDC/.ERSST/.version5/.sst/T/(Jan%201980)/last/RANGE/Y/-5/5/RANGEEDGES/X/190/240/RANGEEDGES%5BX/Y%5Daverage/zlev/removeGRID/dods'
n34 = xr.open_dataset(n34_url, decode_times=False)
n34['T'].attrs['calendar'] = '360_day'
n34 = xr.decode_cf(n34, decode_times=True)
n34.sst.plot()
```
