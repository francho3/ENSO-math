# ENSO in a nutshell
"Once about every four years, the *sea surface temperature* (SST) in the eastern Pacific increases by a few degrees over a period of about one year. These events are called **El Niño** (literally: the little boy), referring to the Christmas Child, since the maximum of the event is usually around December. This phenomenon arises through large scale interaction between the Pacific ocean and the overlying atmosphere and is associated with variations in the *equatorial surface winds*. Normally, the equatorial Pacific surface winds, the trade winds, are directed westward and are driven by a pressure difference between a high pressure region in the east (e.g., at Tahiti) and a low pressure region in the west (e.g., at Darwin). During an El Niño, the pressure is lower than normal in Tahiti and higher than normal in Darwin. These variations in *atmospheric pressure* are known as the **Southern Oscillation**. (Philander, 1990). The El Niño/Southern Oscillation (ENSO) phenomenon is the most prominent interannual signal in the climate system and has large effects on the weather, even far outside the Pacific basin." Dijkstra, (2000).

Our story begins in the port of Paita in northern Peru. 

For most of the year, the coastal water is cold and the current flows northward. In (northern hemisphere) winter, the water warms. Fishermen since at least the 19th century had called this warm current the current of *'El Niño'* (the Christ child) because of its appearance around Christmas.

In some years, this current is more noticeable and is accompanied by unusual climate conditions. 1891 was such an El Niño year. In his 1895 address to the Sixth International Geographical Congress, Senor Federico Alfonso Pezet of the Lima Geographical Society said,

Key points
* Southern hemisphere summer = northern hemisphere winter
* Counter-current flowing southward, contrary its usual northward course
* Observed near the coast
* Large impacts on the local temperature and rainfall
    * Heavy rain in normally desert coastal area
    

The climate (or **climatology**) refers to the conditions that happens every year on average.  
Anomalous conditions (or **anomalies**) are the difference between the climatology and what happens.  
Climate = average of weather

Climatology is what you expect, anomalies are what you get.
- Climatology = "normal"
- Anomaly = departure from "normal"

If the temperature $X$ is considered to be a random variable, $\mu_X = E[X]$ is its climatological mean and $X' = X - \mu_X$ is the anomaly.

Climatology = climatological mean

Because temperature has an annual cycle, we often compute its statistics separately for each month of the year.

For instance, $\mu_{X_t|t \in \text{Jan}} = E[X_t | t \in \text{Jan}]$ would be the January climatology.

Meanwhile about a decade earlier in 1877 there had been a failure of the Indian Monsoon.  
There was drought and famine (Great famine of 1876-78).  
The death toll was estimated at 5.5 million.

Henry Blanford was the first head in 1875 of what came to be the India Meteorological Department (IMD).

He was asked to make monsoon forecasts, and he did, basing his forecasts on the thickness and extent of Himalayan snow. This was an observed association (correlation). In fact, correlation was just being invented. Galton, F. (1890).

$$ \text{cor}(x,y) = \frac{\text{cov}(x,y)}{\sigma_x \sigma_y} $$

Sir Gilbert Walker was the 2nd IMD director (1904-24). Walker was aware of a seesaw of atmospheric pressure across the Indo-Pacific region, which he call the Southern Oscillation.

```python
IGY_sst_url = 'https://iridl.ldeo.columbia.edu/SOURCES/.NOAA/.NCDC/.ERSST/.version5/.anom/T/3/runningAverage/T/%28Dec-Feb%201957%29/VALUE/zlev/removeGRID/Y/-30/30/RANGE/' + t_fix + 'dods'
IGY_sst_ds = xr.open_dataset(IGY_sst_url)

proj = ccrs.PlateCarree(central_longitude=180)
fig = plt.figure(figsize=(25,4))
ax  = plt.subplot(projection=proj)

IGY_sst_ds.anom.isel(T=0).plot.contourf(transform=ccrs.PlateCarree(),vmin=-2, levels=9)
ax.coastlines();
plt.title('Dec 1957 - Feb 1958 SST anomaly (departure from normal)');
```

A good rule of thumb, especially in the tropical Pacific, is  
- persistent precipitation is found over the warmest water
- in such regions, the net motion is upward
- the surface pressure is low.

<div style=float:left>
<img src="images/Walker_Neutral_large.jpg" width="400"/><img src="images/AnnualSST.png" width="450"/>
</div>

The Walker circulation.
1. The eastern equatorial Pacific sea surface temperature (SST) is relative cool, and the western Pacific SST is relatively warm.
2. This temperature difference results in a thermal circulation
    1. Warm moist air rises in the west (low pressure)
    2. Cool dry air descends in the east (high pressure)
    3. Westerly winds aloft, easterly winds at the surface (trade winds) from high to low pressure
    5. The easterly winds drive ocean upwelling (Ekman transport) and cool SST in the east
3. This ocean-atmosphere coupling is relevant for the annual mean, seasonal cycle, and ENSO.

"The greatest advances since Bjerknes' day have been on the oceanographic side of the problem. Although the oceanographic component of his scenario is not very specific, he did recognize that the variations in SST during El Niño are related to ocean dynamics and not to changes in surface heat flux. The key to these dynamics was to switch attention from SST to sea-level variations. The work of Wyrtki (1975, 1979) in collecting and interpreting sea-level data from a network of tide gauges in the tropical Pacific is the basis of our present understanding of the oceanography of El Niño."

<div style="text-align: right"><small> Cane, M. A. (1986). El Niño. Annual Review of Earth and Planetary Sciences, 14(1), 43-70.</a> </div>

# 4. First sucessful El Niño forecast

The first model to simulate realistic ENSO variability was the Zebiak–Cane model (1987), which was constructed by coupling the tropical atmosphere model of Zebiak (1985) to the upper-ocean model of Zebiak & Cane (1985). This model simulated anomalies about prescribed climatological surface winds, SST, and ocean currents.

"If the winds alter the SST pattern, which alters the winds,
then a coupled ocean-atmosphere model is clearly required."

Ability to model/simulate $\neq$ ability to predict. Why not? Examples?
- See chaos, Lorenz, butterfly effect, extreme sensitivity to initial conditions

Nonetheless, the first successful El Niño forecast was made in 1986
